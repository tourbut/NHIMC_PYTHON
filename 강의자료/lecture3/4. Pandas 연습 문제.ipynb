## 2. 인덱스 설정
`일자` 컬럼을 인덱스로 설정해주세요.

일자 인덱스를 datetime 형식으로 변경한 뒤, 날짜 오름차 순으로 데이터를 정렬할 것입니다.  
astype() 메서드를 사용해서 sample.index를 str형으로 바꿔주세요.

참고) https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.astype.html

```python
#일자 컬럼 타입 확인
sample.dtypes
```

```python
sample['일자'] = sample['일자'].astype("str")
```

str로 변경된 인덱스를 pd.to_datetime 메서드를 사용하여 datetime 형식으로 바꿔주세요.

참고) https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html

```python
#답변 작성
sample['일자'] = pd.to_datetime(sample['일자'])
```

```python
#확인
sample['일자']
```

```python
#답변 작성
sample = sample.set_index("일자")
```

sort_index() 메서드를 사용하여 주어진 데이터를 인덱스 날짜 오름차 순으로 정렬해주세요.

참고) https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_index.html

```python
#답변 작성
sample = sample.sort_index(ascending=True)
```

## 4. apply 활용

각 행에 대해 남성 인구의 합을 구하여 `남자인구 합계` 이름을 가진 컬럼을 생성해주세요.

```python
def getManPopulation(row):
    man = 0
    
    for i in row.index:
        if i.startswith("남자"):
            man += row[i]
    
    #답변 작성
            
    return man

sample['남자인구 합계'] = sample.apply(getManPopulation, axis=1)

sample.head()
```

```python
sample.drop('남자인구 합계', axis=1, inplace=True)
```

```python
남자_컬럼들 = [col for col in sample.columns if '남자' in col]
```

```python
sample['합계2'] = sample[남자_컬럼들].sum(axis=1)
```

## 5. 시각화

남성인구 중 남자0\~9세가 차지하는 비율을 계산하여 "남자0~9세 비율" 컬럼을 만들어주세요.

```python
#답변 작성
sample['남자0~9세 비율'] = sample['남자0~9세'] / sample['남자인구 합계'] * 100
```

시간의 흐름에 따라 전체 남자 생활인구 대비 `남자0~9세 비율`의 변화를 한 눈에 알 수 있도록  
`남자0~9세 비율`을 선 그래프로 나타내주세요.


* 한글 폰트가 깨지지 않도록 처리해주세요.
* 그래프 사이즈는 20\*10으로 해주세요.

```python
import matplotlib.pyplot as plt

plt.rc('font', family='malgun gothic')
plt.figure(figsize=(20,10))

#답변 작성
sample['남자0~9세 비율'].plot(kind='line',title='남자0~9세 비율 추이',xlabel='일자',ylabel='비율')
```

```python
# sample 데이터프레임에서 '남자_컬럼들'에 해당하는 컬럼만 선택하여 첫번째 행을 가져온 후, 
# pie chart로 시각화합니다.
# kind='pie' : pie chart로 시각화합니다.
# figsize=(10,10) : 그래프의 크기를 설정합니다.
# title='2020-01-01 남자 연령대별 비율' : 그래프의 제목을 설정합니다.
# autopct='%.2f%%' : 각 항목의 비율을 표시합니다. 소수점 둘째자리까지 표시합니다.
# fontsize=7 : 폰트 크기를 설정합니다.
# legend=True : 범례를 표시합니다.
# labels=남자_컬럼들 : 각 항목의 라벨을 설정합니다.
# labeldistance=1.1 : 라벨과 그래프의 중심점 사이의 거리를 설정합니다.
# startangle=90 : 그래프의 시작 각도를 설정합니다. 90도는 12시 방향입니다.

indexing = 2

sample[남자_컬럼들].iloc[indexing].plot(kind='pie'
                                    , figsize=(10,10)
                                    , title=sample.index[indexing].strftime("%Y-%m-%d")+' 남자 연령대별 비율', autopct='%.2f%%'
                                    , fontsize=7
                                    , legend=True
                                    , labels=남자_컬럼들
                                    , labeldistance=1.1
                                    , startangle=90
                                    )
```
